import needed libraries and modules

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

# Function to load and display the original image

```python
def display_image(img, title="Image"):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis("off")
    plt.show()
```

# Function to perform K-Means clustering and return segmented image

```python

def segment_image_kmeans(img, K):
    pixel_values = img.reshape((-1, 3))
    pixel_values = np.float32(pixel_values)
    kmeans = KMeans(n_clusters=K)
    labels = kmeans.fit_predict(pixel_values)
    centers = np.uint8(kmeans.cluster_centers_)
    segmented_image = centers[labels.flatten()]
    segmented_image = segmented_image.reshape(img.shape)
    return segmented_image
```

# Load the image

```python
image_path = "mikasa.webp"
img = cv2.imread(image_path)
print("Original Image:")
display_image(img, title="Original Image")
```

displaying the segmented images


```python
K_values = [3, 5, 10]

for K in K_values:
    segmented_img = segment_image_kmeans(img, K)
    print(f"Segmented Image with K={K}:")
    display_image(segmented_img, title=f"Segmented Image with K={K}")

```
